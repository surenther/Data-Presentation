The objective of this project is to communicate the systemic collapse of childcare affordability in the United States between 2008 and 2018. Based on an analysis of over 34,000 records, the data reveals that the average American family now spends 16.39% of their income on infant care—well over double the federal affordability benchmark of 7%.

To effectively reach our target audience of State Legislators and Labor Departments, the communication strategy is divided into three distinct, non-overlapping mediums:

* The Interactive Dashboard: Focuses on geographic disparity, allowing policy advisors to identify high-burden "hotspots" and compare provider types (Center vs. Family-based) to guide regional funding.

* The Executive Economic Report: Focuses on macroeconomic trends, utilizing high-density visuals to correlate rising childcare costs with suppressed workforce participation and stagnant wage growth.

* The "Childcare Anchor" Presentation: Focuses on narrative impact, humanizing the data through longitudinal visuals that show the cumulative financial burden on families as children age.

By utilizing these three unique platforms, the project ensures that decision-makers receive both the granular data needed for policy adjustment and the compelling narrative required for public advocacy.

### Medium 1: Interactive Geographic Dashboard

<b> Target Audience: </b> Federal and State Labor Departments
<b> Narrative: </b> This dashboard provides a localized view of the childcare crisis, allowing policy advisors to identify specific "hotspots" where the affordability gap is most severe. By visualizing data geographically, the dashboard helps move beyond national averages to show where specific interventions or subsidy adjustments are needed most. The interactive nature allows users to filter by provider type to see if family-based care offers a viable, lower-cost alternative to centers in high-burden counties.

<b> Mockup Layout: </b>

* Visual A: The "Gap" Choropleth Map: A US map shaded by county. The shading indicates the percentage points a county is above the 7% HHS affordability threshold (e.g., dark red indicates counties >15% above the target).

* Visual B: Provider Price Comparison (Bar Chart): A side-by-side bar chart showing MCInfant (Center) vs. MFCCInfant (Family-based) prices, categorized by state.

* Visual C: High-Burden Leaderboard (Metric List): A ranked list of the top 10 most expensive counties based on the Infant_Affordability_Ratio, highlighting the 54.3% maximum found in the data.

* Visual D: Year Scrubber (Slider): A timeline slider (2008–2018) that allows the user to watch the map "redden" as costs outpace income over the decade.

### Medium 2: Executive Economic Report

<b> Target Audience: </b> Legislative Budget Committees
<b> Narrative: </b> This report frames childcare as a macroeconomic infrastructure issue. It focuses on the "Scissors Effect"—the widening divergence between household wage growth and childcare inflation. By connecting these costs to Female Labor Force Participation (FLFPR), the report provides the evidence needed to argue that affordable childcare is a prerequisite for labor market stability and regional economic growth.

<b> Mockup Layout: </b>

* Visual A: The "Scissors" Graph (Line Chart): A dual-axis chart. The primary axis shows MHI (Income) growth, while the secondary axis shows the much steeper growth curve of MCPreschool prices from 2008 to 2018.

* Visual B: Workforce Correlation (Scatter Plot): Plotting FLFPR_20to64 on the Y-axis and Infant_Affordability_Ratio on the X-axis to visualize the negative correlation between high costs and workforce participation.

* Visual C: The "Net-Pay" Waterfall Chart: A chart starting with average Gross Income ($46,432), with segments "falling away" for estimated taxes and average childcare costs ($7,594), ending with a small bar representing remaining disposable income.

* Visual D: Statistical Heat Table: A formatted table with color-coded cells showing the mean, median, and standard deviation of costs across different child age groups.

### Medium 3: "The Childcare Anchor" Presentation

<b> Target Audience: </b> State Legislators and Community Stakeholders
<b> Narrative: </b> This presentation humanizes the dataset of 34,567 records for a broader audience. It follows the "life of a family" through a decade of care. The visuals are designed to be high-impact for a slide deck, using clear, large-scale imagery to demonstrate how childcare costs "anchor" families to lower economic mobility and prevent them from contributing to the local economy.

<b> Storyboard Agenda (Slide by Slide): </b>

* Slide 1: The Affordability Gauge (Dial Chart): A series of three "speedometer" gauges showing the 7% HHS Goal vs. the 16.39% National Average vs. the 54.30% Maximum Burden.

* Slide 2: The Age Escalator (Stacked Area Chart): A chart showing the cumulative financial burden as a single child moves from Infant to Toddler to Preschooler, illustrating that the "financial hit" stays high for years.

* Slide 3: Geographic Inequity (Treemap): A treemap where the size of the squares represents the TotalPop (Population) living in counties categorized by their burden levels (e.g., 7% goal, 10-15% burden, and >20% crisis zones).

* Slide 4: Regional Variance (Box-and-Whisker Plot): A plot showing the distribution of childcare prices across all counties in a state, highlighting how "zip code lottery" affects a family's financial health.

* Slide 5: Policy Action (Summary Graphic): An icon-based graphic summarizing the three key data-driven recommendations (e.g., subsidy expansion, tax credits, and provider support).
